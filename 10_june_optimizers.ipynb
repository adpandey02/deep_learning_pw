```python
'''
Part 1: Role of Optimization Algorithms in Artificial Neural Networks

Optimization algorithms play a crucial role in training artificial neural networks. Their primary purpose is to minimize the loss function of a neural network by iteratively adjusting the network's parameters. By optimizing the parameters, the algorithm aims to find the optimal set of weights and biases that yield the best performance for a given task.
Optimization algorithms are necessary for several reasons:
Convergence: Neural networks are typically trained using an iterative optimization process. Optimization algorithms ensure that the model converges to an optimal solution by finding the parameters that minimize the loss function. Without optimization algorithms, finding the optimal set of parameters would be a challenging and time-consuming task.
Generalization: Neural networks need to generalize well to unseen data. Optimization algorithms help prevent overfitting by finding parameter values that strike a balance between fitting the training data and capturing the underlying patterns in the data. This improves the network's ability to make accurate predictions on new, unseen examples.
Efficiency: Optimization algorithms provide efficient ways to compute the gradients of the loss function with respect to the network's parameters. These gradients are essential for updating the parameters during the training process. By efficiently calculating and applying these updates, optimization algorithms enable faster convergence and more efficient training of neural networks.

Gradient Descent and its Variants:
Gradient descent is a fundamental optimization algorithm used in neural networks. It operates by iteratively adjusting the network's parameters in the opposite direction of the gradient of the loss function with respect to the parameters. The goal is to reach the minimum of the loss function.
There are different variants of gradient descent, including:
Batch Gradient Descent (BGD): BGD computes the gradient of the loss function by considering the entire training dataset at once. It updates the parameters after evaluating the gradient on the entire dataset. BGD provides accurate gradient estimates but can be computationally expensive for large datasets.
Stochastic Gradient Descent (SGD): SGD computes the gradient of the loss function by considering only a single training example at each iteration. It updates the parameters immediately after processing each example. SGD is computationally efficient but can be noisy due to the high variance in the gradients.
Mini-Batch Gradient Descent: Mini-batch GD falls between BGD and SGD. It computes the gradient by considering a small subset or a mini-batch of training examples. Mini-batch GD strikes a balance between accuracy and computational efficiency.
The tradeoffs between these variants lie in terms of convergence speed and memory requirements. BGD tends to converge slowly as it computes the gradients on the entire dataset but requires more memory. SGD converges faster due to more frequent updates but can exhibit more oscillations. Mini-batch GD combines the advantages of both BGD and SGD, offering a compromise between accuracy and efficiency.

Challenges and Modern Optimizers:
Traditional gradient descent optimization methods face several challenges. Some of these challenges include slow convergence, getting stuck in local minima, and the difficulty of choosing appropriate learning rates.
Modern optimizers address these challenges in various ways. For example:
Momentum: Momentum is a concept introduced in optimization algorithms to accelerate convergence. It adds a fraction of the previous update vector to the current update, allowing the optimizer to continue progressing even in flat or narrow regions of the loss surface. Momentum helps overcome slow convergence by dampening oscillations and speeding up convergence in the relevant direction.
Learning Rate: The learning rate determines the step size taken by the optimizer during parameter updates. Choosing an appropriate learning rate is crucial, as a too high or too low value can hinder convergence. Modern optimizers often employ adaptive learning rate techniques that adjust the learning rate dynamically based on the progress of the optimization. This adaptive behavior helps optimize the convergence speed and prevents overshooting or getting stuck in localminima.



Part 2: Optimizer Techniques

Stochastic Gradient Descent (SGD):
Stochastic Gradient Descent (SGD) is a variant of gradient descent that computes the gradient of the loss function using a single training example at each iteration. SGD has several advantages compared to traditional gradient descent methods:
Computational Efficiency: SGD updates the parameters after processing each training example, making it computationally efficient, especially for large datasets.
Avoiding Local Minima: SGD's noisy updates can help escape local minima. The noise introduced by considering a single example at a time can prevent the optimizer from getting stuck in flat regions of the loss surface.
However, SGD also has limitations. The high variance of the gradient estimates can lead to a slow convergence rate and oscillations during training. Additionally, SGD requires careful tuning of the learning rate, as choosing an inappropriate value can lead to convergence issues.

Adam Optimizer:
Adam (Adaptive Moment Estimation) is an optimizer that combines the concepts of momentum and adaptive learning rates. It maintains a running average of past gradients and their squared values to adaptively adjust the learning rate for each parameter.
Benefits of Adam optimizer include:
Adaptive Learning Rates: Adam adjusts the learning rates for each parameter based on the magnitudes of the gradients. It automatically adapts the learning rate, providing faster convergence and reducing the need for manual tuning.
Momentum: Adam incorporates momentum by using the first-order moments (mean) of the gradients. It helps dampen oscillations and accelerates convergence, especially in the presence of sparse gradients.
One potential drawback of Adam is that it requires more memory to store the additional running average parameters compared to other optimizers.

RMSprop Optimizer:
RMSprop (Root Mean Square Propagation) is an optimizer that addresses the challenges of adaptive learning rates. It calculates the adaptive learning rate for each parameter based on the exponentially weighted average of past squared gradients.
RMSprop has the following advantages:
Adaptive Learning Rates: Similar to Adam, RMSprop adjusts the learning rates for each parameter. It adapts the learning rate based on the magnitude of recent gradients, providing faster convergence and stable updates.
Memory Efficiency: RMSprop requires less memory compared to Adam since it does not store the running average of the past gradients.
Comparing RMSprop to Adam, RMSprop has lower memory requirements but may have slower convergence in some cases. Adam, on the other hand, offers faster convergence but at the cost of higher memory usage'''
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Load and preprocess the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train / 255.0
x_test = x_test / 255.0

# Build the model
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))

# Compile the model with SGD optimizer
sgd_optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
model.compile(optimizer=sgd_optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model with SGD optimizer
sgd_history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

# Compile the model with Adam optimizer
adam_optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=adam_optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model with Adam optimizer
adam_history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

# Compile the model with RMSprop optimizer
rmsprop_optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.001)
model.compile(optimizer=rmsprop_optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model with RMSprop optimizer
rmsprop_history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

import matplotlib.pyplot as plt

# Plot learning curves (loss)
plt.figure(figsize=(12, 6))
plt.plot(sgd_history.history['loss'], label='SGD')
plt.plot(adam_history.history['loss'], label='Adam')
plt.plot(rmsprop_history.history['loss'], label='RMSprop')
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot learning curves (accuracy)
plt.figure(figsize=(12, 6))
plt.plot(sgd_history.history['accuracy'], label='SGD')
plt.plot(adam_history.history['accuracy'], label='Adam')
plt.plot(rmsprop_history.history['accuracy'], label='RMSprop')
plt.title('Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Evaluate model performance on the test set
sgd_loss, sgd_accuracy = model.evaluate(x_test, y_test)
adam_loss, adam_accuracy = model.evaluate(x_test, y_test)
rmsprop_loss, rmsprop_accuracy = model.evaluate(x_test, y_test)

print("Test Set Performance:")
print("SGD: Loss =", sgd_loss, " Accuracy =", sgd_accuracy)
print("Adam: Loss =", adam_loss, " Accuracy =", adam_accuracy)
print("RMSprop: Loss =", rmsprop_loss, " Accuracy =", rmsprop_accuracy)
```
